## Custom : top 10 passeurs

```python
import seaborn as sns

top_assisters = df.groupby('Player', observed=True)['Assists'].sum() \
                 .sort_values(ascending=False).head(10).sort_values()

fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(top_assisters.index, top_assisters.values,
        color=sns.color_palette('Blues_r', len(top_assisters)), edgecolor='black')
ax.set_title("Top 10 des passeurs décisifs (2017-2024)")
ax.set_xlabel('Passes décisives totales')
for i, v in enumerate(top_assisters.values):
    ax.text(v + 1, i, f'{int(v)}', va='center', fontweight='bold')
plt.show()
```
