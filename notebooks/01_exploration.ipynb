## Chargement (format européen : ; et ,)

```python
df = load_raw()
print(f'Shape : {df.shape}')
df.head()
```

```python
print_section('Ligues')
print(df['league'].value_counts())

print_section('Saisons')
print(sorted(df['season'].unique()))

print_section('Dimensions')
print(f'Joueurs uniques : {df["player"].nunique():,}')
print(f'Équipes         : {df["team"].nunique()}')
print(f'Nations         : {df["nation_"].nunique()}')
print(f'Positions       : {df["pos_"].nunique()}')
```

## Stats clés

```python
key_cols = ['age_', 'Playing Time_MP', 'Playing Time_Min',
            'Performance_Gls', 'Performance_Ast', 'Expected_xG']
df[key_cols].describe().round(2)
```

## Missing values (par catégorie)

```python
missing = df.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)
print(f'Total colonnes avec missing : {len(missing)}')
print(f'Missing total : {missing.sum():,}')
missing.head(20)
```

## Distribution des buts

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
df['Performance_Gls'].hist(bins=30, ax=axes[0], edgecolor='black')
axes[0].set_title('Distribution des buts par joueur·saison'); axes[0].set_xlabel('Buts')

df.loc[df['Performance_Gls'] >= 10, 'Performance_Gls'].hist(bins=30, ax=axes[1], edgecolor='black', color='orange')
axes[1].set_title('Buteurs à 10+ buts')
plt.tight_layout(); plt.show()
```
